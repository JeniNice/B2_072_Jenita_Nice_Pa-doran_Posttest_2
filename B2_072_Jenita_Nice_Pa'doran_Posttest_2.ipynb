```python
with open("dataset_market_basket.txt") as file: #membuka file "dataset_market_basket.txt" dan menampilkannya sebagai "file", menggunakan "with" statement untuk memastikan bahwa file ditutup secara otomatis setelah selesai membaca.
  raw_data = file.read() #membaca seluruh konten dari file yang disimpan dalam objek "file", dan menempatkannya ke dalam variabel "raw_data".
```

```python
raw_data #isi dari file txt yang sudah dibaca dan disimpan 
```

```python
data = raw_data.split("\n") #memecah string "raw_data" menjadi beberapa string yang terpisah berdasarkan karakter newline "\n" dan menyimpannya dalam bentuk list yang disimpan di dalam variabel "data"
```

```python
dataset =[item.split(",") for item in data] #engubah setiap string dalam list "data" menjadi list terpisah yang masing-masing merepresentasikan item dalam transaksi, dengan menggunakan koma sebagai karakter pemisah.
```

```python
#melakukan penggabungan semua list dalam variabel "dataset" menjadi satu list tunggal, dan menyimpannya dalam variabel "items"
items = [item for sublist in dataset for item in sublist]

#membuat set unik dari items
unique_items = set(items)

# menghitung frekuensi setiap jenis item pada setiap baris dalam dataset
tabular_data = []
for row in dataset:
    row_data = {}
    for item in unique_items:
        row_data[item] = row.count(item)
    tabular_data.append(row_data)    

# membuat tabel tabular menggunakan pandas DataFrame
df = pd.DataFrame(tabular_data)
df
```

```python
#ubah dataset menjadi bentuk tabulasi 
col = set()
for trc in dataset:
  for item in trc:
    col.add(item)

#List Kolom
col
len(col)
```

```python
data = []
for data in dataset:
  kolom_tabulasi = []
  for item in col:
    biner = 0
    for item_data in data:
      if item_data == item :
        biner = 1

      kolom_tabulasi.append(biner)
    
    data.append(kolom_tabulasi)

data
```

```python
from itertools import permutations

rules = list(permutations(df.columns,2)) #menghasilkan semua kemungkinan urutan dua kolom dalam data frame
rules
```

```python
hitung = []
for rule in rules:
  banyak_data = len(df[
      (df[rule[0]]==1) & (df[rule[1]]==1)
  ])

  hitung.append(banyak_data)

hitung
```

```python
len(df)
```

```python
rule_baru = [] # Menyimpan data dalam bentuk list
for rule in rules: # Mengulangi pembacaan dataset
  if len(df[(df[rule[0]]==1) & (df[rule[1]]==1)]) >0: # Menentukan apakah kombinasi tersebut ada atau tidak
    rule_baru.append(rule) # Menyimpan hasil pengecekan dalam bentuk tuple didalam list
rule_baru
```

```python
hitung = {} # Menyimpan data dalam bentuk dictionary
for rule in rule_baru:  # Mengulangi pembacaan dataset
  banyak_data= len(df[
      (df[rule[0]]==1) & (df[rule[1]]==1) # Menentukan berapa jumlah kemunculan kombinasi dan menyimpannya dalam sebuah variabel
      ])
  hitung[str(rule)] = banyak_data # Menambahkan hasil penentuan ke dalam dictionary
hitung 
```

```python
# List support

for item,ab in hitung.items(): # Untuk melakukan perulangan pembacaan dataset pada tiap - tiap kombinasi itemnya
  print(f"{item} support = {ab/len(df)*100}%") 
```

```python
# List Confidens

for item,ab in hitung.items(): # Untuk melakukan perulangan pembacaan dataset pada tiap - tiap kombinasi itemnya
  #"A->B"
  i = eval(item) #Ubah ke list dari string
  print(f"{i[0]}->{i[1]}  Confidence = {ab/df[i[0]].sum()*100}%")
```

```python
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.neighbors import KNeighborsClassifier
```

```python
df2 = pd.read_csv("deteksi_uang_palsu.csv")
df2
```

### **1. Menambahkan kolom jarak untuk menampilkan jarak dari dataset terhadap target prediksi**

```python
# Membuat tabel jarak 


# membuat dataframe dari data baru
new_data = pd.DataFrame({'diagonal'    : [172.25],
                         'height_left' : [103.94],
                         'height_right': [103.68],
                         'margin_low'  : [4.14],
                         'margin_up'   : [3.25],
                         'length'      : [112.48]})

# Menghitung jarak antara setiap baris dataframe utama dengan dataframe data baru
df2['distance'] = euclidean_distances(df2[['diagonal', 'height_left', 'height_right', 'margin_low', 'margin_up', 'length']], new_data)

# Menampilkan tabel jarak
print('Tabel Jarak : ')
print(df2['distance'])
```

```python
# Menambah kolom jarak
pd.set_option('display.max_columns', None)

# Mengatur lebar tabel menjadi 200
pd.set_option('display.width', 200)

print(df2)
```

### **2. Mengurutkan data berdasarkan K jarak terdekat (Untuk K nya ambil akar 2 dari banyak data)**

```python
# Mengurutkan data dari kecil ke besar
sorted_data = df2.sort_values('distance')

# Menentukan nilai k
k = int(np.sqrt(len(df2)))

# Menentukan data terdekat
data_terdekat = sorted_data.head(k)

# Menampilkan data terdekat
print(data_terdekat)
```

### **3. Menentukan hasil prediksinya berdasarkan label terbanyak**

```python
# Split data menjadi data training dan data testing
X_train = df2.drop(columns=['is_genuine','distance'],axis=1)
y_train = df2['is_genuine']

# inisiasi data baru menjadi data tes
test_data = new_data

# membuat variabel knn yang berisikan algoritma KNN dengan nilai K = akar 2 dari banyak data
knn = KNeighborsClassifier(n_neighbors=k)

# melatih knn dengan data training
knn.fit(X_train, y_train)

# memprediksi hasil
prediksi = knn.predict(test_data)

print("Berdasarkan penggunaan library sklearn untuk memprediksi menggunakan Algoritma KNN, hasilnya adalah : "+ ' '.join(map(str, prediksi)))
```

Dataset "Insurance" memiliki 7 kolom dengan penjelasan sebagai berikut:

* age: Umur pelanggan dalam tahun
* sex: Jenis kelamin pelanggan, dengan nilai 0 untuk wanita dan 1 untuk pria
* bmi: Indeks massa tubuh pelanggan
* children: Jumlah anak yang dimiliki oleh pelanggan
* smoker: Status merokok pelanggan, dengan nilai 0 untuk tidak merokok dan 1 untuk merokok
* region: Wilayah tempat tinggal pelanggan, dengan nilai 'northeast', 'southeast', 'southwest', atau 'northwest'
* charges: Biaya premi yang harus dibayar oleh pelanggan dalam satu tahun
Kolom "age", "sex", "bmi", "children", dan "smoker" adalah variabel independen yang digunakan untuk menjelaskan variabel dependen "charges".

Kolom "age", "sex", "bmi", dan "children" adalah variabel numerik yang dapat dianalisis secara kuantitatif, sedangkan kolom "smoker" dan "region" adalah variabel kategorikal yang harus diubah menjadi variabel dummy untuk analisis lebih lanjut.

Kolom "charges" adalah variabel dependen yang menjadi fokus analisis dataset ini. Variabel ini merepresentasikan biaya premi yang harus dibayar oleh pelanggan dalam satu tahun. Variabel dependen ini dapat dipengaruhi oleh variabel independen, dan dapat digunakan untuk mengembangkan model prediksi biaya premi asuransi.
